## Classifying Files Using Multinomial Naive Bayes Classification ##

In this exercise, we'll be classifying text documents into 1 of twenty different categories using multinomial naive bayes classification. Our bayes classification model scans through each document and predicts which topic that document belongs to based on its contents. 

For this exercise, we’ll be using the 20 Newsgroups data set. There are several versions of it on
the web but we will use the "20news-bydate.tar.gz” from
http://qwone.com/~jason/20Newsgroups/
Overall, there are roughly 19,000 documents, each from one of 20 newsgroups. The label of a document is the identity of its
newsgroup. The documents are divided into a training set and a test set.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import itertools
import scipy
import pandas as pd
import sklearn
from sklearn import metrics
from sklearn import datasets
from sklearn.datasets import fetch_20newsgroups
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
traindata = sklearn.datasets.load_files('20news-bydate/20news-bydate-train/' )
testdata = sklearn.datasets.load_files('20news-bydate/20news-bydate-test/')
```

```python
testdatamap = dict(enumerate(testdata.filenames))
traindatamap = dict(enumerate(traindata.filenames))
reversetestdatamap = {(v.split('/')[10] +'/'+ v.split('/')[11]): k  for k, v in testdatamap.items()}
```

```python
## create dictionary from testdata.target_names as a mapping
categories = {}
for i,e in enumerate(testdata.target_names):
    categories [i] = e
```

## Building the vocabulary ##

We also need the “vocabulary.txt” file

```python
vocab = {}
reversevocab = {}
count = 0
with open("vocabulary.txt",'r') as voc:
    for i in voc:
        j = i.strip()
        vocab[j] = count
        reversevocab[count] = j
        count +=1
```

## Creating the Count vectorizer and Multinomial Naive Bayes model using sklearn ##

We create the count vectorizer with strip_accents and decode error parameters to prevent unicode errors in the count vectorization. Additionally, we use the english stop words to filter commonly used words and increase accuracy

```python
count_vect = CountVectorizer(strip_accents= 'unicode', decode_error= 'ignore',stop_words= 'english',vocabulary=vocab)
training_counts = count_vect.fit_transform(traindata.data)
training_feature_names = count_vect.get_feature_names()

tfidf_transformer = TfidfTransformer()
training_tfidf = tfidf_transformer.fit_transform(training_counts)
print(training_tfidf.shape)

clf = MultinomialNB(alpha=1.0).fit(training_tfidf, traindata.target)

test_counts = count_vect.transform(testdata.data)
test_tfidf = tfidf_transformer.transform(test_counts)

predicted = clf.predict(test_tfidf)
# predicted
```

## Building the model##
The words in the documents constitute an overall vocabulary V of size 61188. Build a multinomial
Naive Bayes model using the training data. For each of the 20 classes j = 1, 2, . . . , 20, we need the following:

• π<sub>j</sub> , the fraction of documents that belong to that class; and

• P<sub>j</sub> , a probability distribution over V that models the documents of that class.


## Creating π<sub>j</sub> , the fraction of documents that belong to that class##

Using the known targets for the raw training data, we can tally up the total number of files belonging to each class and use this to find the percentage of all files belonging to each class.

Sklearn also allows using the naive_bayes.MultinomialNB class to output all the counts of each classes with clf.class_count()

```python
# unique,counts = numpy.unique(predicted, return_counts = True)
# counts_predicted = dict(zip(unique,counts))
# print(counts_predicted)
# print("Total files in testing data: {}".format(numpy.sum(counts_predicted.values())))

unique, counts = np.unique(traindata.target,return_counts = True)
counts_traindata = dict(zip(unique,counts))
num_counts_traindata = np.sum(counts_traindata.values())
print(counts_traindata) 
print("Total files in training data: {}").format(num_counts_traindata)
```

```python
# totalfiles = training_tfidf.shape[0] + test_tfidf.shape[0] ## total number of files. 
## should be 11314 training + 7532 testing = 18846

print('πj , the percentage of documents that belong to that class:' + '\n')
summ = 0
for i in range(20):
    documentnum = counts_traindata[i]*100.000
    documentfraction = documentnum/num_counts_traindata
    summ = summ + documentfraction
    print(categories[i] +  ': {0:.3f}%').format(documentfraction)
print("Total: {}%").format(summ)
print("Total Number of files: {}").format(num_counts_traindata)

```

## P<sub>j</sub> , a probability distribution over V that models the documents of that class##

For each wordw ∈ V , let Pjw be the fraction of this concatenated document occupied by w

"clf.feature\_count_" returns a weighted distribution of all the words over each class which we can pass into a pandas dataframe using the feature names as columns and target names as the index

```python
countdf = pd.DataFrame(clf.feature_count_, columns= training_feature_names,index= traindata.target_names)
countdf.head()
```

We can find the percentages by summing each row and dividing 

```python
summs = list(countdf.sum(axis=1))
unweighted = pd.DataFrame()
for i in range(len(countdf.index)):
    unweighted[categories[i]] = (countdf.loc[categories[i]] / summs[i] * 100)
unweighted = unweighted.transpose()
unweighted
```

## Distribution of Predictions over all 20 classes ##

Additionally, we can see the distribution of the model's orediction over all 20 classes

```python
proba = clf.predict_proba(test_tfidf)
probdist = {}
for i in range(20):
    probdist[i] = 0
    for j in proba:
        probdist[i] = probdist[i] + j[i]
        
probdist
for i in range(20):
    probfrac = probdist[i]*100/test_tfidf.shape[0] ## divide by number of files
    print(categories[i] +  ': {0:.3f}%').format(probfrac)
print("Total: {}%").format(summ)
```

## Writing a routine ## 

Wrapping the code into a fuction that uses this naive Bayes model to classify a new document. To avoid underflow,
work with logs rather than multiplying together probabilities.

This function outputs the probability of a file in each class given the category and file name as a string

```python
def probdistfunc (directory_and_name):
    filenumber = reversetestdatamap[directory_and_name]
    filename = str(testdatamap[filenumber])
    print('Probability of file {} in class:' + '\n').format(filename)
    proba1 = clf.predict_proba(test_tfidf[filenumber])
    for i in range(20):
        probpercentage = proba1[0][i] * 100.000
        print(categories[i] +  ': {0:.3f}%').format(probpercentage)
    print("Total: {}%").format(summ)
```

## Evaluate the performance of your model on the test data ##

Using the array given by the prediction model, we find our accuracy is around 82%

```python
print(metrics.classification_report(testdata.target, predicted))
Accuracy = metrics.accuracy_score(testdata.target, predicted)
print("Accuracy = " + str(Accuracy))
print('Error Rate = ' + str(1-Accuracy))
```

Which we can show in a confusion matrix

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    #plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)

#     if normalize:
#         cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
#         print("Normalized confusion matrix")
#     else:
#         print('Confusion matrix, without normalization')

#         print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
cnf_matrix = metrics.confusion_matrix(testdata.target, predicted)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure(figsize = (8,8))
plot_confusion_matrix(cnf_matrix, classes=traindata.target_names,
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
# plt.figure()
# plot_confusion_matrix(cnf_matrix, classes=traindata.target_names, normalize=True,
#                       title='Normalized confusion matrix')

plt.show()
```
